```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Map the species integer to the species name
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_df['species'] = iris_df['species'].map(species_map)

# Check the data types of each column
print("Data types before conversion:")
print(iris_df.dtypes)

# Convert all feature columns to numeric if they are not already
iris_df[iris.feature_names] = iris_df[iris.feature_names].apply(pd.to_numeric)

# Check the data types again after conversion
print("\nData types after conversion:")
print(iris_df.dtypes)

# Pairplot
sns.pairplot(iris_df, hue='species')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()

# Boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(data=iris_df, orient="h", palette="Set2")
plt.title("Boxplot of Iris Features")
plt.show()

# Violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='sepal length (cm)', data=iris_df)
plt.title("Violin Plot of Sepal Length by Species")
plt.show()

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=iris_df, palette='Set1', s=100)
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.show()
```
